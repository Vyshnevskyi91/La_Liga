```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.skysports.com/la-liga-table'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
header = soup.find_all('th')
```

```python
header = [i.text for i in header]
```

```python
df = pd.DataFrame(columns=header)
```

```python
df
```

```python
text = soup.find_all('tr')
```

```python
for j in text[1:]:
    row_data = j.find_all('td')
    cell = [x.text.strip() for x in row_data]
    length = len(df)
    df.loc[length] = cell
```

```python
df = df.set_index('#')
```

```python
df.drop('Last 6', axis=1, inplace=True)
```
